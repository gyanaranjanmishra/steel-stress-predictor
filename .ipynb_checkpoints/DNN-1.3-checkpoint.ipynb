```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('DATA/1.3Si-Flow.csv')
```

```python
X=df.drop('stress',axis=1).values
y=df['stress'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.30, random_state=100)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense,Dropout
```

```python
from tensorflow.keras.constraints import max_norm
```

```python
from tensorflow.keras.models import load_model
```

```python
model = Sequential()

# input layer
model.add(Dense(30,  activation='relu'))


# hidden layer
model.add(Dense(15, activation='relu'))


# hidden layer
model.add(Dense(15, activation='relu'))


# output layer
model.add(Dense(units=1))

# Compile model
model.compile(loss='mean_absolute_error', optimizer='adam')
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=150,
          batch_size=64,
          validation_data=(X_test, y_test), 
          )
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
w1= pd.DataFrame(model.layers[3].get_weights()[1])
w1.to_csv('Lb4.csv', index=False)
```

```python
losses.plot()
```

```python
losses.to_csv('losses-13.csv', index=False)
```

```python
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test,predictions))
```
